
# Exploratory Data Analysis

## Pie Chart Of Gender Overview

```python
image_folder = "D:\Data Sceince\Prodigy\Task 1\Visuals"
```

```python
explode=(0.1,0)
fig1, ax1 = plt.subplots(figsize=(12,5))
ax1.pie(df["Gender"].value_counts(),explode=explode, labels=["Male","Female"],
        autopct="%1.1f%%",shadow=True, startangle=90)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis("equal")
plt.tight_layout()
plt.legend()
plt.title("Gender Overview", fontsize=20)
plt.savefig(os.path.join(image_folder,"Gender Overview"))
plt.show()
```

## Distribution of Employee Ages

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
sns.histplot(data=df, x=df["Age"], bins=15, kde=True, color="green")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.title("Distribution of Employee Ages", fontsize=14)
plt.savefig(os.path.join(image_folder,"Distribution Of Employee Ages"))
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(15,5))
sns.countplot(data=df,x="BusinessTravel",hue="Attrition",palette="coolwarm")
plt.xlabel("Bussiness Travel Frequency", fontsize=12)
plt.ylabel("Number of Employees",fontsize=12)
plt.title("Employee Attrition Based on Business Travel", fontsize=14)
plt.legend(title="Attrition", labels=["No","Yes"])
plt.savefig(os.path.join(image_folder,"Employee Attrition Based On Business Travel"))
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(15,7))
sns.countplot(data=df,x="EducationField",hue="Attrition",palette="bright")
plt.xlabel("Education Field",fontsize=12)
plt.ylabel("Number Of Employees",fontsize=12)
plt.title("Employees Attrition Based On Education Field")
plt.legend()
plt.savefig(os.path.join(image_folder,"Employee Attrition Based On Education Field"))
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(15,7))
sns.countplot(data=df,x="Department",hue="Attrition",palette="colorblind")
plt.xlabel("Department",fontsize=12)
plt.ylabel("Number Of Employees",fontsize=12)
plt.title("Employees Attrition Based On Department")
plt.savefig(os.path.join(image_folder,"Employee Attrition Based On Department"))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df,x="Age",hue="Attrition",bins=15,kde=True,palette="coolwarm",alpha=0.6)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Distribution of Age by Attrition", fontsize=14)
plt.savefig(os.path.join(image_folder,"Distribution Of Age By Attrition"))
plt.show()
```
